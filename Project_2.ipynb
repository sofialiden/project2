```python
import pandas as pd
url= "https://raw.githubusercontent.com/KadeF10/Spotify/main/data.csv"
df= pd.read_csv(url)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.regplot(x="danceability", y="tempo", data=df)
plt.ylim(0,)
```
